```python
import bs4
import requests
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import os
import nltk
import nltk.corpus

#web scrapping
page_url = "https://www.livemint.com/Companies/BLvAa2vOFbCZoua8UpIliI/CBI-charges-Vijay-Mallya-with-fraud-criminal-conspiracy.html"
uClient = uReq(page_url)

page_soup = soup(uClient.read(), "html.parser")
uClient.close()

containers = page_soup.findAll("div",{"class":"contentSec"})
for x in containers:
    ans = x.find('p')
scrapped_text=ans.text
scrapped_text
#---------------------

#tokenization and a basic check
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 

punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*’“_~'''
for x in scrapped_text:
    if x in punctuations:
        scrapped_text_1 = scrapped_text.replace(x,"") 
  
stop_words = set(stopwords.words('english')) 
  
word_tokens = word_tokenize(scrapped_text_1) 
filtered_sentence = [] 
for w in word_tokens: 
    
    if w not in stop_words: 
        filtered_sentence.append(w) 
  
finalstr = ' '.join(map(str, filtered_sentence)) 
print(finalstr)
print('\n')

scrapped_tokens = word_tokenize(finalstr)
scrapped_tokens.sort()
sc_tkn1 = []
for item in scrapped_tokens:
    sc_tkn1.append(item.lower())
y = set(sc_tkn1)

#my keyword inventory
list1 = ['scam','robbery','terrorism','weapons','war','breach','malicious','fake','murder','kill','arrested','criminal','conspiracy',
         'judicial','custody','laundering','investigation','fraud','owed','charge','accused','arrests','debt',
         'raids','charged','corruption','penal','misutilisation','misappropriation','penalty','non-bailable','jail']

list3 = []
list3 = [x for x in list1 if x in sc_tkn1]
print("my keyword inventory--->",list1,'\n')
print("matched words--->",list3,'\n')
is_match = len(list3)/len(y)*100
print("% of negative word match--->","%.2f" % is_match,"%",'\n')


#using VADER Sentiment Analysis

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
count = 1
count2=0
s2 = ""
for w in scrapped_text: 
    if w!=".":
        s2+=w
    elif w == '.':
        if len(s2) >50:
            print("sentence ",count," : ",s2,end=" ")
            print('\n')
            sid_obj = SentimentIntensityAnalyzer()
            sentiment_dict = sid_obj.polarity_scores(s2)
            print("Overall sentiment dictionary is : ", sentiment_dict) 
            print("sentence was rated as ", sentiment_dict['neg']*100, "% Negative") 
            print("sentence was rated as ", sentiment_dict['neu']*100, "% Neutral") 
            print("sentence was rated as ", sentiment_dict['pos']*100, "% Positive")
        
            print("Sentence Overall Rated As", end = " ") 
        
  
    # decide sentiment as positive, negative and neutral 
            if sentiment_dict['compound'] >= 0.05 : 
                print("Positive",'\n') 
  
            elif sentiment_dict['compound'] <= -0.05 : 
                print("Negative",'\n')
                count2+=1
  
            else : 
                print("Neutral",'\n') 
            count+=1
            s2=""
print("The number of negative sentences are :",count2)       
```
